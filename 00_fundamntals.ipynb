Random and pesudo random tensors

```python
# we use random datasets to get tensors with zero bias which can then be changed during training the model

random_1 = tf.random.Generator.from_seed(3) # global seed
# initialize (from_seed) the Generator at a value means to show the list of an alrealdy existing random list from said index value
random_1 = random_1.normal(shape=(3,2))

random_2 = tf.random.Generator.from_seed(3)
random_2 = random_2.normal(shape=(3,2))

print(random_1)
print(random_2)

print(random_1 == random_2) # since Generator (index) is same, we get the same list both times
```

```python
# shuffling the order of elements (useful, so that inherent order dosent affect the training)
'''
Let's say you working with 15,000 images of cats and dogs and the first 10,000 images of were of cats
and the next 5,000 were of dogs.
This order could effect how a neural network learns (it may overfit by learning the order of the data),
instead, it might be a good idea to move your data around.
'''
```

```python
not_shuffled = tf.constant([[10,7],[5,2],[3,4]])
shuffled = tf.random.shuffle(not_shuffled) # only shuffles along 1st dimension
print(shuffled)
```

```python
'''
tf.random.shuffle() relies on a combination of two different seeds: the Global Seed and the Operation Seed.
When you only provide seed=42 inside the function, you are setting the Operation Seed.
However, if the Global Seed is not set, TensorFlow picks a new one every time you run the code, which changes the final outcome.
'''

shuffled = tf.random.shuffle(not_shuffled, seed = 42)
print(shuffled)
```

```python
'''
TensorFlow's rule for randomness works like this:

-If you set neither: You get a different random result every time.
-If you set the Global Seed only: You get a predictable sequence of random numbers.
-If you set the Operation Seed only (seed=42): TensorFlow pairs it with a shifting global counter.
 Since that counter changes every time you execute the cell, your "shuffle" changes too.
'''

tf.random.set_seed(42) # sets global seed
shuffled = tf.random.shuffle(not_shuffled, seed = 42)
print(shuffled)
```

Turn numpy arrays into tensors (for faster gpu computation)

```python
'''
Note: A matrix or tensor is typically represented by a capital letter
(e.g. X or A) where as a vector is typically represented by a lowercase letter (e.g. y or b
'''
import numpy as np
numpy_A = np.arange(1,25,dtype=np.int32) # create an arrafy ranging from 1-25
print(numpy_A)

tensor_A = tf.constant(numpy_A) # creates a tensor of numpy_A in the shape (24,0,0)
print(tensor_A)
```

```python
tensor_A = tf.constant(numpy_A, shape = (2,4,3)) # creates the tensor in a particular shape from the array
print(tensor_A)
```

Getting info from tensors

```python
rank_4_tensor = tf.zeros(shape = [2,3,4,5])
print(rank_4_tensor)
```

```python
print("Datatype of every element: ", rank_4_tensor.dtype)
print("Number of dimensions: ", rank_4_tensor.ndim)
print("Shape: ", rank_4_tensor.shape)
print("Elements along axis-0: ", rank_4_tensor.shape[0])
print("Elements along last axis: ", rank_4_tensor.shape[-1])
print("Number of elements: ", tf.size(rank_4_tensor))
print("Number of elements: ", tf.size(rank_4_tensor).numpy()) # gives direct value as an array
```

```python
# tensor indexes similar to py

rank_4_tensor[:2,:2,:2,:2] # gives the first 2 items of each dimension
```

```python
rank_4_tensor[:1,:1,:1,:] # gets the first element from each dim except the last
```

```python
# add an extra dimension (at the end)

rank_2_tensor = tf.constant([[10,7],[3,4]])
rank_3_tensor = rank_2_tensor[..., tf.newaxis] # in Python "..." means "all dimensions prior to"
print(rank_3_tensor)
print(rank_2_tensor.shape,rank_3_tensor.shape)
```

```python
# or we can also expand the dimension by
tf.expand_dims(rank_2_tensor, axis = -1) # to expand on last axis
```
